```python

import dis

def find_ycap(x,w):
    y_cap = []
    for i in x:
        y_cap.append(i*w)
    return y_cap

def error(y,y_cap,e):
    for i in range(len(y)):
        e += (y[i] - y_cap[i])**2
    #e = [(y[i] - y_cap[i])**2 for i in range(len(y))]
    Error = e/len(y) 
    #Error = [i/len(y) for i in e]
    print(Error)    
    return Error
            
def weight(w,Error,lr):
    wc = w - (-(Error*lr))  
    print(wc)
    return wc
    
def main():
    x = [1,2,3,4]
    y = [2,4,6,8]
    w = 0.8
    wc,e,Error = 0,0,0
    Error_weight_dict = {}
    lr =  0.1
    for i in range(10):
        y_cap = find_ycap(x,w).copy()
        print(f"new y cap values : {y_cap}")
        Error = error(y,y_cap,e)
        w = weight(w,Error,lr)
        Error_weight_dict[Error] = w
    print(Error_weight_dict)
    print("min error--->", min(Error_weight_dict))
    print("opt weight---->", Error_weight_dict[min(Error_weight_dict)])

    
if __name__ == "__main__":
    main()
```

```python
dis.dis(main)
```

```python
import math
x = [1,2,3,4]
y = [2,4,6,8]
n = len(x)
w = 0.8
b = 0
alpha = 0.1
iterations = 10

def hyp(x, w, b):
    y_hat=[]
    for xi in x:
        y_hat.append((w*xi) + b)
    return y_hat

def cost(y, y_hat):
    e = [(y[i] - y_hat[i]) for i in range(n)]; print('e--->', e)
    sq_error = [ (i**2) for i in e]; print('sqr_e--->', sq_error)
    E = sum(sq_error)/n
    return E

def gd(w, E):
    w = w - (-(E*alpha))    
    return w

def regression(x, y, w, b):
    E_w={}

    for i in range(iterations):
        print(i+1)    
        y_hat = hyp(x, w, b); print('y_hat--->', y_hat)
        E = cost(y, y_hat); print ('E--->', E)
        w = gd(w, E); print ('w--->', w)
        E_w[E]=w
        print("-------------------");print(f"E_w[E] : {E_w[E]}")

    print(E_w)
    print("")
    print("min error--->", min(E_w))
    print("opt weight---->", E_w[min(E_w)])

regression(x, y, w, b)
```

```python
dis.dis(regression)
```
